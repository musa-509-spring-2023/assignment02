```python
import geopandas as gpd
import pandas as pd
import sqlalchemy as sqa
from geoalchemy2 import Geography

df = gpd.read_file('tl_2020_42_bg/tl_2020_42_bg.shp')

df.rename_geometry('geog', inplace=True)

df = df.to_crs(epsg=4326)

df
```

```python
USERNAME = 'postgres'
PASSWORD = 'postgres'
DATABASE = 'assignment_2'
```

```python
engine = sqa.create_engine(
    f'postgresql://{USERNAME}:{PASSWORD}@localhost:5432/{DATABASE}'
)
df.to_postgis(
    'blockgroups_2020',
    engine,
    dtype={'geog': Geography()},
    schema='census',
    if_exists='replace',
    index=False,
)
```

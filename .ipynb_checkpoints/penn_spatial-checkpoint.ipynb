```python
import geopandas as gpd
import pandas as pd
import sqlalchemy as sqa
from geoalchemy2 import Geography

df = gpd.read_file('__data__/penn/ef7af340-1817-41f6-ad5b-032f11d6d7872020328-1-54q7hz.msruy.shp')

df.rename_geometry('geog', inplace=True)

df = df.to_crs(epsg=4326)



studyarea = [
    "University of Pennsylvania"
]

df = df.query("NAME in @studyarea")



df
```

```python
USERNAME = 'postgres'
PASSWORD = 'postgres'
DATABASE = 'assignment_2'
```

```python
engine = sqa.create_engine(
    f'postgresql://{USERNAME}:{PASSWORD}@localhost:5432/{DATABASE}'
)
df.to_postgis(
    'penn',
    engine,
    dtype={'geog': Geography()},
    schema='phl',
    if_exists='replace',
    index=False,
)
```
